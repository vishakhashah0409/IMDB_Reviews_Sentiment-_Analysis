# Bag of words

```python
def get_bag_of_words(X_train, X_test, X_valid):
    # Load the train, test, and validation data
    train_data = pd.read_csv('X_train.csv')
    test_data = pd.read_csv('X_test.csv')
    valid_data = pd.read_csv('X_valid.csv')
    
    # Create CountVectorizer object
    vectorizer = CountVectorizer()
    
    #Fit the vectorizer on the train data
    vectorizer.fit(train_data['processed'])
    
    # Transform the train, test, and validation data
    train_bow = vectorizer.transform(train_data['processed'])
    test_bow = vectorizer.transform(test_data['processed'])
    valid_bow = vectorizer.transform(valid_data['processed'])
    
  
    # Print the shape of the transformed data
    print("Shape of the transformed data:")
    print("Train set: ", train_bow.shape)
    print("Test set: ", test_bow.shape)
    print("Validation set: ", valid_bow.shape)
    
    # Print the bag of words
    print("Bag of words (vocabulary):")
    print(vectorizer.vocabulary_)
    print("")
    
    
    # Print the size of the vocabulary and the feature shape
    print("Vocabulary size:", len(vectorizer.vocabulary_))
    print("Train bow shape:", train_bow.shape)
    
    # Return the train, test, and validation features
    bow = {'train': train_bow, 'test': test_bow, 'valid': valid_bow}
    return bow

# Call the function to extract BoW features
features = get_bag_of_words(train_data['processed'], test_data['processed'], valid_data['processed'])
```

# 1)  Multinomial Naive Bayes model

```python
# Create a multinomial naive Bayes model
mnb_model = MultinomialNB()

# Fit the model on the train data
mnb_model.fit(train_bow, Y_train)
mnb_model.fit(train_tfidf, Y_train)


# Make predictions on the test data
mnb_bow_preds = mnb_model.predict(test_bow)
mnb_tfidf_preds = mnb_model.predict(test_tfidf)


# Evaluate the model on the test set
mnb_bow_acc = accuracy_score(Y_test, mnb_bow_preds)
mnb_tfidf_acc = accuracy_score(Y_test, mnb_tfidf_preds)

print("Multinomial Naive Bayes bow Accuracy: {:.2f}% ".format(mnb_bow_acc * 100))
print("Multinomial Naive Bayes tfidf Accuracy: {:.2f}% ".format(mnb_tfidf_acc * 100))

```

```python
#create confusion matrix
bow_matrix = confusion_matrix(Y_test, mnb_bow_preds)
tfidf_matrix = confusion_matrix(Y_test, mnb_tfidf_preds)

print(bow_matrix)
print(tfidf_matrix)


print(classification_report(Y_test, mnb_bow_preds))
print(classification_report(Y_test, mnb_tfidf_preds))
```

```python
#heatmap
print(sns.heatmap(bow_matrix/np.sum(bow_matrix), annot=True,fmt='.2%'))
```

# 3) Support vector machine

# Hyperparameter Tuning on best model

```python

lr_model = LogisticRegression()
# Define the parameter grid to search over
param_grid = {
    'C': [0.1, 1.0, 10.0],
    'penalty': ['l1', 'l2']
}

# Create a grid search object and fit it to the training data
grid_search = GridSearchCV(lr_model, param_grid, cv=5)
grid_search.fit(train_tfidf, Y_train)

# Print the best hyperparameters and the corresponding cross-validation score
print('Best hyperparameters: {}'.format(grid_search.best_params_))
print('Cross-validation score: {:.2f}'.format(grid_search.best_score_))
```

```python
# Evaluate the accuracy of the best estimator on the testing set
best_estimator = grid_search.best_estimator_
lr_model = LogisticRegression()
lr_model.fit(train_tfidf, Y_train)
lr_tfidf_preds = best_estimator.predict(test_tfidf)
lr_tfidf_acc = accuracy_score(Y_test, lr_tfidf_preds)
print("Logistic Regression tfidf Accuracy: {:.2f}% ".format(lr_tfidf_acc * 100))
#confusion matrix
tfidf_matrix=confusion_matrix(Y_test, lr_tfidf_preds)
print(tfidf_matrix)
print("\n")
print(classification_report(Y_test, lr_tfidf_preds))
```

```python
# Evaluate the accuracy of the best estimator on the testing set
best_estimator = grid_search.best_estimator_
lr_model = LogisticRegression()
lr_model.fit(train_tfidf, Y_train)
val_preds_lr = best_estimator.predict(valid_tfidf)
val_acc_lr = accuracy_score(Y_valid, val_preds_lr)
print("Validation accuracy for Logistic Regression: {:.2f}% ".format(val_acc_lr*100))
```
