```python
data1=pd.read_csv("C:/Users/visha/OneDrive/Documents/processeddata.csv")
data1
```

```python
# Binarisation of Sentiments: Positive: 1, Negative: 0

lb=LabelBinarizer()

#Transformed sentiment data
sentiment_data=lb.fit_transform(data1['sentiment'])
print(sentiment_data.shape)
```

```python
#train dataset
X = data1.processed
y = data1.sentiment
print(X.shape, y.shape)

data1.isnull().sum()
```

```python
#split the dataset 
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(X,y,test_size = 0.3,random_state = 45)
     
```

```python
#bag of words
from sklearn.feature_extraction.text import CountVectorizer
vectorizer = CountVectorizer()
X_train = vectorizer.fit_transform(X_train)
X_test = vectorizer.transform(X_test)
X_train.shape


featureName = vectorizer.get_feature_names()
print("Number of features - ", len(featureName))
print("Last 30 features are\n", featureName[-30:])


vectorizer.vocabulary_
```

```python
from scipy import stats

from keras.datasets import imdb
from keras_preprocessing.sequence import pad_sequences
from keras.models import Sequential
from keras.layers.embeddings import Embedding
from keras.layers import LSTM, Dense, Activation
from keras.preprocessing.text import Tokenizer
```

```python
tokenizer = Tokenizer(num_words=5000, split=' ')
tokenizer.fit_on_texts(data1['processed'].values)
X = tokenizer.texts_to_sequences(data1['processed'].values)
X = pad_sequences(X_train)
```

```python
# Creating the LSTM model
model = Sequential()
model.add(Embedding(5000, 128, input_length=X.shape[1]))
model.add(LSTM(128, dropout=0.2, recurrent_dropout=0.2))
model.add(Dense(1, activation='sigmoid'))
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
print(model.summary())
```

```python
# Training the model
model.fit(X, data1['sentiment'], epochs=4, batch_size=32, verbose=1)
```

```python
# Testing the model
test = ['The movie was really bad, I did not like it at all',
        'The movie was fantastic, I loved every bit of it']
test = tokenizer.texts_to_sequences(test)
test = pad_sequences(test, maxlen=X.shape[1])
print(model.predict(test))
```

```python
loss, accuracy = model.evaluate(X, data1['sentiment'], verbose=0)
print("Training: accuracy = %f  ;  loss = %f" % (accuracy, loss))
```

```python
score, acc = model.evaluate(X, data1['sentiment'], batch_size=32)

print('Test score:', score)
print('Test accuracy:', acc)
```
